# Theoretical description



I describe here a simple 2-site/2-electron model that is suitable for demonstrating the 
Floquet idea

Hamiltonian:
\begin{equation}
H = \epsilon \sum_{i=1,2}\sum_\sigma a^{\dagger}_{i\sigma}a_{i\sigma}
+ t \sum_\sigma (a^{\dagger}_{1\sigma}a_{2\sigma}+a^{\dagger}_{2\sigma}a_{1\sigma})
+U\sum_{i=1,2}(n_{i\uparrow}-1/2)(n_{i\downarrow}-1/2)
\end{equation}
Following the usual convention, we take $t < 0$ so that delocalizating the
 electron over the two sites is energetically favored. 

For this system, we can define the symmetrized basis:
\begin{eqnarray}
|\phi_1 \rangle = a^{\dagger}_{1\uparrow}a^{\dagger}_{2\downarrow}|0\rangle\\
|\phi_2 \rangle = a^{\dagger}_{2\uparrow}a^{\dagger}_{1\downarrow}|0\rangle\\
|\phi_3 \rangle = a^{\dagger}_{1\uparrow}a^{\dagger}_{1\downarrow}|0\rangle\\
|\phi_4 \rangle = a^{\dagger}_{2\uparrow}a^{\dagger}_{2\downarrow}|0\rangle\\
|\phi_5 \rangle = a^{\dagger}_{1\uparrow}a^{\dagger}_{2\uparrow}|0\rangle\\
|\phi_6 \rangle = a^{\dagger}_{1\downarrow}a^{\dagger}_{2\downarrow}|0\rangle
\end{eqnarray}

+ $\phi_1$ and $\phi_2$ denote covalent configurations,
+ $\phi_3$ and $\phi_4$ are ionic configurations with 
both electrons, spin paired, on a single site,
+  $\phi_5$ and $\phi_6$ are triplet configurations.



Next, we define the symmetrized states:
\begin{eqnarray}
\phi_{ag} = \frac{1}{\sqrt{2}}(\phi_1  + \phi_2)\\
\phi_{bg} = \frac{1}{\sqrt{2}}(\phi_3  + \phi_4)\\
\phi_{au} = \frac{1}{\sqrt{2}}(\phi_1  - \phi_2)\\
\phi_{bu} = \frac{1}{\sqrt{2}}(\phi_3  - \phi_4)
\end{eqnarray}
and the transformation $T = \{\phi_{ag},\phi_{bg},\phi_{au},\phi_{bu},\phi_5,\phi_6\}$, 
then
\begin{eqnarray}
\tilde H = T\cdot H \cdot T^{\dagger}
\end{eqnarray}
is block-diagonal within the $2\times 2$ block spanned by 
$\{\phi_{ag},\phi_{bg}\}$ (which is spin singlet).  
The remaining 
states are eigenstates with energies $2\epsilon$ for $\{\phi_{au},\phi_5,\phi_6\}$ corresponding to the three
triplet spin combinations and $2\epsilon + U$ for the $\phi_{bu}$, which is a spin singlet compose of an anti-symmetric combination of the two possible
ionic configurations. 

The final $2\times 2$ matrix can be brought into diagonal form by 
defining 
\begin{eqnarray}
 \phi_{g+} = c \phi_{ag} + s \phi_{bg}\\
 \phi_{g-} = -s \phi_{ag} + c \phi_{bg}
 \end{eqnarray}
 where  $c^2 + s^2 = 1$ with corresponding eigenvalues
 \begin{eqnarray}
 \epsilon_{g\pm} = 2 \epsilon + \frac{1}{2}(U \mp\sqrt{16t^2+U^2})
\end{eqnarray}
 The ground state, is then the $\phi_{g+}$ state with energy
$$ \epsilon_{g+} = 2 \epsilon + \frac{1}{2}(U -\sqrt{16t^2+U^2})$$

Note that $c$ and $s$ are cosine and sine functions of the mixing angle
defined by 
\begin{eqnarray}
\tan(\theta) = -\frac{U}{2t} 
\end{eqnarray}

The ratio $U/t$ controls the localization of the eigenstates.  

* $U/t \ll 1$ the delocalized/covalent
$\phi_{ag}$ combination.  

* $U/t \gg 1$  the ground state
is dominated $\phi_{bg}$ consisting of the two ionic configurations.  
  
## Definition of transition dipole operator
For the Floquet analysis, we need to define the a transition operator that couples the spring eigenstates.  Since optical transitions preserve spin, all of the Floquet driven transitions must be within the spin-singlet manifold. 
Two transitions are allowed:
$\phi_{g+}\to \phi_{bu}$ and $\phi_{bu}\to \phi_{g-}$, thus we define 2 transition 
operators
\begin{eqnarray}
\hat\mu_1  &=& \mu_{ge} \left(|\phi_{g+}\rangle \langle \phi_{bu}| + |\phi_{bu}\rangle \langle \phi_{g+}|\right) \\ 
\hat\mu_2  &=& \mu_{e2e} \left(|\phi_{g-}\rangle \langle \phi_{bu}| + |\phi_{bu}\rangle \langle \phi_{g-}|\right) 
\end{eqnarray}
where $\mu_{ge}$ and $\mu_{e2e}$ are the values of the transition moments. 

## reduced 3-state model
Thus, we arrive at a reduced 3-state model
\begin{eqnarray}
H(t) &=& \sum_{i=1,3} \epsilon_i |\psi_i\rangle\langle \psi_i|
\nonumber 
\\
&+& \mu_1\cos(\Omega t)  (|\psi_1\rangle\langle \psi_2| + |\psi_2\rangle\langle \psi_1|)\nonumber \\
&+& \mu_2 \cos(\Omega t ) (|\psi_3\rangle\langle \psi_2| + |\psi_2\rangle\langle \psi_3|)
\end{eqnarray}
where $\Omega$ is the angular frequency of the external driving field, $\mu_1$ and $\mu_2$ are
the couplings to the field, and 
\begin{eqnarray}
\epsilon_1 &=&  \frac{1}{2}(U -\sqrt{16t^2+U^2}) \\
\epsilon_2 &=& U \\
\epsilon_3 &=& \frac{1}{2}(U +\sqrt{16t^2+U^2})
\end{eqnarray}
Notice that the system has 2 distinct frequencies controlled entirely by the electron
repulsion term. 
If $U = 0$, the $1\to 2$ and $2\to 3$ energy gaps are identical.  


In the implementation below, I take $\mu_1 = \mu_2 = g/2$ for simplicity.
This parameter we vary from small to large at a fixed values of the
parameters t,U, and $\Omega$.  The energy scale is set to be $U =1$

```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
```

```python

psi1  = basis(3, 0)
psi2 = basis(3, 1)
psi3 = basis(3, 2)
#print(psi1, psi2, psi3)
u = 1. # U sets the energy scale of the system.
t =   u
delta = (1/2)*(np.sqrt(16*t**2+u**2))
e1 = (u/2)-delta
e2 = u
e3 = (u/2)+delta
omega = 2*np.pi
H0 = e1*psi1*psi1.dag()+e2*psi2*psi2.dag()+e3*psi3*psi3.dag()

dipole = (psi1*psi2.dag()+psi2*psi1.dag()) + (psi2*psi3.dag()+psi3*psi2.dag())
#A = np.linspace(0, 100, 100)

H1 = dipole/2.0

args = {'w': omega}

H = [H0, [H1, 'sin(w * t)']]
T = 2*np.pi / omega
f_modes_0, f_energies = floquet_modes(H, T, args)
print(f_modes_0)
print(f_energies)
```

In this case we have u & t fixed and scan the coupling. 
u = 1 and t = 1.

note that the floquet propagation needs to occur over exactly 1 period of the external driving field.

need a plot of the coefficients for each mode as a function of the coupling.

```python
g_vec = np.linspace(0, 100, 100)*u #* omega;
T = (2*np.pi)/omega
tlist  = np.linspace(0.0, 10 * T, 101)
psi0 = psi1
q_energies = np.zeros((len(g_vec), 3))
args = {'w': omega}
for idx, g in enumerate(g_vec):
  H1 = g/2.0 * dipole
  H = [H0, [H1, lambda t, args: np.sin(args['w']*t)]]
  f_modes, f_energies = floquet_modes(H, T, args, True)
  q_energies[idx,:] = f_energies


plt.figure(figsize=(14,8))
plt.plot(g_vec/u, q_energies[:,0] / delta, 'b', g_vec/u, q_energies[:,1] / delta, 'r',  g_vec/u, q_energies[:,2] / delta, 'g')
plt.xlabel(r'$g/U$')
plt.ylabel(r'Quasienergy / $U$')
plt.title(r'Floquet quasienergies')
plt.show()
```
